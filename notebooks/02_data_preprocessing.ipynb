# Phase 2: Data Preprocessing

## 🎯 Goals:
1. Load raw datasets
2. Handle missing values (mean imputation)
3. Remove Insulin & SkinThickness from diabetes dataset (6 features total)
4. Split into training/testing sets (80/20)
5. Save clean data (NO missing values)
6. **DO NOT create scalers** (Phase 3 will do this)

## ⚠️ Important:
- We do NOT scale data here
- We do NOT create scalers here
- We ONLY clean and split the data
- Scalers will be created in Phase 3!

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
```

## 1. Load Raw Data

```python
# Load datasets
diabetes = pd.read_csv('../data/raw/diabetes.csv')
hypertension = pd.read_csv('../data/raw/hypertension.csv')

print("📊 Diabetes Dataset:")
print(f"Shape: {diabetes.shape}")
print(f"Columns: {diabetes.columns.tolist()}")

print("\n📊 Hypertension Dataset:")
print(f"Shape: {hypertension.shape}")
print(f"Columns: {hypertension.columns.tolist()}")
```

## 2. Handle Missing Values - Diabetes

### Strategy:
1. Replace impossible zeros with NaN
2. Fill NaN with column mean
3. **Remove Insulin and SkinThickness columns** (we'll use 6 features)

```python
print("🔧 Processing Diabetes Dataset...")
print("="*80)

# Check for zeros (which are actually missing values)
zero_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("\nImpossible zeros found:")
for col in zero_columns:
    zero_count = (diabetes[col] == 0).sum()
    print(f"  {col}: {zero_count} zeros")

# Replace zeros with NaN
diabetes[zero_columns] = diabetes[zero_columns].replace(0, np.nan)

print("\nMissing values after replacing zeros:")
print(diabetes.isnull().sum())

# Fill missing values with mean
print("\n⏳ Filling missing values with column means...")
for col in diabetes.columns:
    if diabetes[col].isnull().sum() > 0:
        mean_val = diabetes[col].mean()
        diabetes[col].fillna(mean_val, inplace=True)
        print(f"  ✓ {col}: filled with mean = {mean_val:.2f}")

# Verify no missing values
print("\n📊 Missing values after imputation:")
print(diabetes.isnull().sum())

if diabetes.isnull().sum().sum() == 0:
    print("\n✅ SUCCESS: No missing values in diabetes dataset!")
else:
    print("\n⚠️ WARNING: Still have missing values!")
```

```python
# Remove Insulin and SkinThickness columns (6-feature model)
print("\n🗑️ Removing Insulin and SkinThickness columns...")
print(f"Before: {diabetes.shape[1]} columns")

diabetes_clean = diabetes.drop(['Insulin', 'SkinThickness'], axis=1)

print(f"After: {diabetes_clean.shape[1]} columns")
print(f"\nRemaining features: {diabetes_clean.columns.tolist()}")
print("\n✅ Now using 6-feature diabetes model (+ Outcome = 7 columns total)")
```

## 3. Handle Missing Values - Hypertension

```python
print("🔧 Processing Hypertension Dataset...")
print("="*80)

# Check for missing values
print("\nMissing values in hypertension dataset:")
missing = hypertension.isnull().sum()
print(missing[missing > 0])

# Fill missing values
print("\n⏳ Filling missing values...")

for col in hypertension.columns:
    if hypertension[col].isnull().sum() > 0:
        # Check if column is numeric or categorical
        if hypertension[col].dtype in ['float64', 'int64']:
            # Numeric: use mean
            mean_val = hypertension[col].mean()
            hypertension[col].fillna(mean_val, inplace=True)
            print(f"  ✓ {col}: filled with mean = {mean_val:.2f}")
        else:
            # Categorical: use mode
            mode_val = hypertension[col].mode()[0]
            hypertension[col].fillna(mode_val, inplace=True)
            print(f"  ✓ {col}: filled with mode = {mode_val}")

# Verify no missing values
print("\n📊 Missing values after imputation:")
print(hypertension.isnull().sum())

if hypertension.isnull().sum().sum() == 0:
    print("\n✅ SUCCESS: No missing values in hypertension dataset!")
else:
    print("\n⚠️ WARNING: Still have missing values!")

hypertension_clean = hypertension.copy()
```

## 4. Split Data (Train/Test)

```python
print("✂️ Splitting datasets into train/test (80/20)...")
print("="*80)

# Split Diabetes (stratified to maintain class balance)
diabetes_train, diabetes_test = train_test_split(
    diabetes_clean,
    test_size=0.2,
    random_state=42,
    stratify=diabetes_clean['Outcome']
)

print("\n📊 Diabetes Split:")
print(f"  Training: {diabetes_train.shape}")
print(f"  Testing:  {diabetes_test.shape}")
print(f"  Features: {diabetes_train.shape[1] - 1}  (excluding Outcome column)")

# Split Hypertension
hypertension_train, hypertension_test = train_test_split(
    hypertension_clean,
    test_size=0.2,
    random_state=42,
    stratify=hypertension_clean['Risk']
)

print("\n📊 Hypertension Split:")
print(f"  Training: {hypertension_train.shape}")
print(f"  Testing:  {hypertension_test.shape}")
print(f"  Features: {hypertension_train.shape[1] - 1}  (excluding Risk column)")
```

## 5. Save Processed Data

```python
import os

# Create processed directory if it doesn't exist
os.makedirs('../data/processed', exist_ok=True)

print("💾 Saving processed datasets...")
print("="*80)

# Save Diabetes
diabetes_train.to_csv('../data/processed/diabetes_train.csv', index=False)
diabetes_test.to_csv('../data/processed/diabetes_test.csv', index=False)
print("✓ Diabetes datasets saved")

# Save Hypertension
hypertension_train.to_csv('../data/processed/hypertension_train.csv', index=False)
hypertension_test.to_csv('../data/processed/hypertension_test.csv', index=False)
print("✓ Hypertension datasets saved")

print("\n📁 Saved files:")
print("  ../data/processed/diabetes_train.csv")
print("  ../data/processed/diabetes_test.csv")
print("  ../data/processed/hypertension_train.csv")
print("  ../data/processed/hypertension_test.csv")
```
